```python
# Install required packages
# !pip install playwright nest-asyncio
# !playwright install

import asyncio
import os

import nest_asyncio
from IPython.display import HTML, Image, display
from playwright.async_api import async_playwright

# Apply nest_asyncio to allow nested event loops
nest_asyncio.apply()


async def load_mhtml_with_control(mhtml_path):
    """
    Load MHTML with more control over the page
    """
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)  # Keep headless for screenshots
        page = await browser.new_page()

        # Set viewport size
        await page.set_viewport_size({"width": 1280, "height": 720})

        # Load the MHTML file
        file_url = f"file://{os.path.abspath(mhtml_path)}"
        await page.goto(file_url, wait_until="networkidle")

        # Wait a bit for any dynamic content
        await page.wait_for_timeout(2000)

        # Get various page information
        info = {
            "title": await page.title(),
            "url": page.url,
            "content": await page.content(),
            "text": await page.text_content("body"),
            "screenshot": await page.screenshot(full_page=True),
            "page": page,
            "browser": browser,
        }

        return info


# Now we can use asyncio.run() with nest_asyncio
mhtml_file = "0daf1895-493d-4b9a-ba8a-ba6a65c23a21_after.mhtml"
page_info = asyncio.run(load_mhtml_with_control(mhtml_file))

print(f"Page Title: {page_info['title']}")
print(f"Page URL: {page_info['url']}")
print(f"Text Content Length: {len(page_info['text'])}")
print(f"HTML Content Length: {len(page_info['content'])}")

# Display the screenshot of the webpage
print("\nWebpage Screenshot:")
display(Image(data=page_info["screenshot"], format="png"))

# Display the HTML content as rendered HTML
print("\nRendered HTML Content:")
display(HTML(page_info["content"]))

# Display first 1000 characters of text content
print("\nText Content Preview:")
print(page_info["text"][:1000])

# Close the browser
await page_info["browser"].close()
```

```python
# Install required packages
# !pip install playwright nest-asyncio
# !playwright install

import asyncio
import random

import nest_asyncio
from IPython.display import HTML, Image, display

# Apply nest_asyncio to allow nested event loops
nest_asyncio.apply()


async def load_mhtml_with_randomization(mhtml_path, randomization_config=None):
    """
    Load MHTML with randomized UI components
    """
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()

        # Set viewport size
        await page.set_viewport_size({"width": 1280, "height": 720})

        # Load the MHTML file
        file_url = f"file://{os.path.abspath(mhtml_path)}"
        await page.goto(file_url, wait_until="networkidle")

        # Wait for content to load
        await page.wait_for_timeout(2000)

        # Apply randomizations
        if randomization_config:
            await apply_randomizations(page, randomization_config)

        # Get page information
        info = {
            "title": await page.title(),
            "url": page.url,
            "content": await page.content(),
            "text": await page.text_content("body"),
            "screenshot": await page.screenshot(full_page=True),
            "page": page,
            "browser": browser,
        }

        return info


async def apply_randomizations(page, config):
    """
    Apply various UI randomizations to the page
    """

    # 1. Randomize colors
    if config.get("randomize_colors", False):
        await randomize_colors(page)

    # 2. Randomize layout positions
    if config.get("randomize_positions", False):
        await randomize_positions(page)

    # 3. Randomize text content
    if config.get("randomize_text", False):
        await randomize_text(page)

    # 4. Randomize element sizes
    if config.get("randomize_sizes", False):
        await randomize_sizes(page)

    # 5. Randomize CSS properties
    if config.get("randomize_css", False):
        await randomize_css(page)

    # 6. Add random elements
    if config.get("add_random_elements", False):
        await add_random_elements(page)


async def randomize_colors(page):
    """Randomize colors of various elements"""
    colors = ["#FF6B6B", "#4ECDC4", "#45B7D1", "#96CEB4", "#FFEAA7", "#DDA0DD", "#98D8C8"]

    # Randomize background colors
    await page.evaluate(f"""
        () => {{
            const elements = document.querySelectorAll('div, p, span, h1, h2, h3, h4, h5, h6, button, input, textarea');
            elements.forEach(el => {{
                if (Math.random() < 0.3) {{
                    el.style.backgroundColor = '{random.choice(colors)}';
                }}
                if (Math.random() < 0.2) {{
                    el.style.color = '{random.choice(colors)}';
                }}
            }});
        }}
    """)


async def randomize_positions(page):
    """Randomize positions of elements"""
    await page.evaluate("""
        () => {
            const elements = document.querySelectorAll('div, p, span, button, input, textarea');
            elements.forEach(el => {
                if (Math.random() < 0.2) {
                    const offsetX = (Math.random() - 0.5) * 100;
                    const offsetY = (Math.random() - 0.5) * 100;
                    el.style.position = 'relative';
                    el.style.left = offsetX + 'px';
                    el.style.top = offsetY + 'px';
                }
            });
        }
    """)


async def randomize_text(page):
    """Randomize text content"""
    random_texts = [
        "Lorem ipsum dolor sit amet",
        "Random text here",
        "Sample content",
        "Test data",
        "Randomized content",
        "Modified text",
        "New content",
    ]

    await page.evaluate(f"""
        () => {{
            const textElements = document.querySelectorAll('p, span, h1, h2, h3, h4, h5, h6, div');
            textElements.forEach(el => {{
                if (Math.random() < 0.1 && el.textContent.trim().length > 0) {{
                    el.textContent = '{random.choice(random_texts)}';
                }}
            }});
        }}
    """)


async def randomize_sizes(page):
    """Randomize sizes of elements"""
    await page.evaluate("""
        () => {
            const elements = document.querySelectorAll('div, p, span, button, input, textarea, img');
            elements.forEach(el => {
                if (Math.random() < 0.15) {
                    const scale = 0.5 + Math.random() * 1.5; // Scale between 0.5 and 2.0
                    el.style.transform = `scale(${scale})`;
                }
                if (Math.random() < 0.1) {
                    const width = 50 + Math.random() * 200;
                    el.style.width = width + 'px';
                }
            });
        }
    """)


async def randomize_css(page):
    """Randomize various CSS properties"""
    await page.evaluate("""
        () => {
            const elements = document.querySelectorAll('*');
            elements.forEach(el => {
                if (Math.random() < 0.05) {
                    // Random border
                    el.style.border = `${Math.random() * 5}px solid #${Math.floor(Math.random()*16777215).toString(16)}`;
                }
                if (Math.random() < 0.03) {
                    // Random opacity
                    el.style.opacity = 0.3 + Math.random() * 0.7;
                }
                if (Math.random() < 0.02) {
                    // Random rotation
                    el.style.transform = `rotate(${(Math.random() - 0.5) * 30}deg)`;
                }
                if (Math.random() < 0.04) {
                    // Random font size
                    el.style.fontSize = (10 + Math.random() * 20) + 'px';
                }
            });
        }
    """)


async def add_random_elements(page):
    """Add random elements to the page"""
    await page.evaluate("""
        () => {
            const body = document.body;
            
            // Add random divs
            for (let i = 0; i < Math.floor(Math.random() * 5) + 1; i++) {
                const div = document.createElement('div');
                div.textContent = 'Random Element ' + (i + 1);
                div.style.position = 'absolute';
                div.style.left = Math.random() * window.innerWidth + 'px';
                div.style.top = Math.random() * window.innerHeight + 'px';
                div.style.backgroundColor = '#' + Math.floor(Math.random()*16777215).toString(16);
                div.style.padding = '10px';
                div.style.border = '1px solid black';
                div.style.zIndex = '9999';
                body.appendChild(div);
            }
            
            // Add random buttons
            for (let i = 0; i < Math.floor(Math.random() * 3) + 1; i++) {
                const button = document.createElement('button');
                button.textContent = 'Random Button ' + (i + 1);
                button.style.position = 'absolute';
                button.style.left = Math.random() * window.innerWidth + 'px';
                button.style.top = Math.random() * window.innerHeight + 'px';
                button.style.zIndex = '9999';
                body.appendChild(button);
            }
        }
    """)


# Configuration for randomizations
randomization_config = {
    "randomize_colors": True,
    "randomize_positions": True,
    "randomize_text": False,  # Set to True if you want to randomize text
    "randomize_sizes": True,
    "randomize_css": True,
    "add_random_elements": True,
}

# Load MHTML with randomizations
mhtml_file = "0daf1895-493d-4b9a-ba8a-ba6a65c23a21_after.mhtml"
page_info = asyncio.run(load_mhtml_with_randomization(mhtml_file, randomization_config))

print(f"Page Title: {page_info['title']}")
print(f"Page URL: {page_info['url']}")

# Display the randomized webpage screenshot
print("\nRandomized Webpage Screenshot:")
display(Image(data=page_info["screenshot"], format="png"))

# Display the modified HTML content
print("\nModified HTML Content:")
display(HTML(page_info["content"]))

# Close the browser
await page_info["browser"].close()
```
